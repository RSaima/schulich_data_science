In this section, we will demonstrated how to calculate descriptive statistics from scratch with Python.

```python
list1 = [1,89,87,65,34,98,87,65]
```

```python
def mean(x):
    return sum(x)/len(x)

mean(list1)
```

```python
def median(x):
    x.sort() # works on any sequential object
    if len(x)%2 == 0:
        med = (x[(len(x)//2)]+x[(len(x)//2)-1])/2 # indexing doesnot accept float therefore // will make it integer
    else:
        med = x[len(x)//2]

    return med

median(list1)
```

```python
def mode(x):
    freq_dict = {}
    for i in x:
        if i in freq_dict:
            freq_dict[i] += 1
        else:
            freq_dict[i] = 1

    # Find the mode(s)
    max_freq = max(freq_dict.values())
    modes = [k for k, v in freq_dict.items() if v == max_freq]
    
    return modes

mode(list1)
```

```python
def iqr(x):
    x.sort()
    first_quartile_val = min(x)
    second_quartile_val = x[len(x)//4]
    third_quartile_val = x[(len(x)//4) * 2]
    fourth_quartile_val = x[(len(x)//4) * 3]
    list_quartiles = [first_quartile_val, second_quartile_val, third_quartile_val, fourth_quartile_val]

    return list_quartiles

iqr(list1)
```

```python
iqr_range = iqr(list1)[-1] - iqr(list1)[1]
print(iqr_range)
```

The following demonstrates the skewness and kurtosis calculations from scratch in Python, and using the Scipy library

```python
# Calculating Kurtosis

import math

def kurtosis(data):
    n = len(data)
    mean = sum(data) / n
    variance = sum((x - mean) ** 2 for x in data) / n
    std_dev = math.sqrt(variance)
    
    kurtosis = sum((x - mean) ** 4 for x in data) / (n * std_dev ** 4) - 3
    return kurtosis
```
